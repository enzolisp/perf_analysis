```python
import numpy as np 
from numpy import linalg
import matplotlib.pyplot as plt
```

Construcao da Malha
 
A malha consiste em uma representação discreta
do domínio [a,b]. Aqui, vamos construir a malha mais simples possível, aquela que
consiste de N pontos igualmente espaçados, isto é, a chamada malha uniforme!

![alt text](image.png)

```python
# construcao da malha
a = 0
b = 1
N = 11
h = (b-a)/(N-1)
x = np.linspace(a,b,N)
```

![alt text](image-3.png)

```python
A = np.zeros((N,N))
b = np.zeros(N)
A[0,0] = 1
b[0] = 0

for i in np.arange(1,N-1):
    A[i,i-1] = 1
    A[i,i] = -2
    A[i,i+1] = 1
    b[i] = -100 * h**2 * (x[i]-1)**2

A[N-1,N-1] = 1
b[N-1] = 0
```

![alt text](image-1.png)

```python
u = np.linalg.solve(A,b)
```

![alt text](image-2.png)

```python
u = np.linalg.solve(A,b)

#def. sol. analitica
def ue(x):
    return -100.0*(x-1)**4/12 - 100*x/12 + 100.0/12
#gráfico
xx = np.linspace(0,1)
yy = np.zeros(50)
for i,xxi in enumerate(xx):
    yy[i] = ue(xxi)
plt.plot(x,u,'ro',xx,yy,'b-')
plt.show()
```
